```python
import pickle as pkl

import matplotlib.pyplot as plt
import seaborn as sns

import numpy as np

import torch
```

```python
cond_samples = pkl.load(open('cond_samples.pkl', 'rb'))
marg_samples = pkl.load(open('marg_samples.pkl', 'rb'))
```

```python
prior_dims = [1, 2, 5, 10, 50, 100, 500, 1000]
```

```python
marg_samples[0].shape
```

```python
plt.scatter(marg_samples[0][:, 0], marg_samples[0][:, 1], label='marginal samples')
plt.scatter(cond_samples[0][:, 0], cond_samples[0][:, 1], label='conditional samples')
```

```python
fig, axs = plt.subplots(2, 4, dpi=300)
for i, ax in enumerate(axs.flatten()):
    ax.scatter(marg_samples[i][:, 0], marg_samples[i][:, 1], label='marginal samples')
    ax.scatter(cond_samples[i][:, 0], cond_samples[i][:, 1], label='conditional samples')
    ax.set_title(f'prior dim = {prior_dims[i]}')
```
