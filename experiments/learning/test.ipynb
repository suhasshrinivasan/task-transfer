```python
import torch
from gensn.variational import VariationalBound
import gensn.distributions as G
import torch.distributions as D
from torch import nn
from gensn.parameters import TransformedParameter
from gensn.variational import VariationalBound
from torch.optim import Adam
```

```python
prior = G.TrainableDistributionAdapter(
    D.Normal, loc=nn.Parameter(torch.Tensor([0.0])), scale=torch.Tensor([2.0])
)
conditional = G.TrainableDistributionAdapter(
    D.Normal,
    loc=nn.Linear(1, 1),
    scale=TransformedParameter(torch.Tensor([1.0]), torch.abs),
)
joint = G.Joint(prior=prior, conditional=conditional)
posterior = G.TrainableDistributionAdapter(
    D.Normal, loc=nn.Parameter(torch.Tensor([0.0])), scale=torch.Tensor([2.0])
)
```

```python
model = VariationalBound(joint=joint, posterior=posterior)
```

```python
optimizer = Adam(model.parameters(), lr=0.01)
```
