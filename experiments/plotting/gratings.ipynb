```python
import numpy as np
from insilico_stimuli.stimuli import GaborSet, PlaidsGratingSet,CenterSurround
import matplotlib.pyplot as plt
```

```python
gabor_params={
    "canvas_size": [12, 12],
    "sizes": [10],
    "spatial_frequencies": [1 / 3],
    "contrasts": [1.0],
    "grey_levels": [0.0],
    "eccentricities": [0.0],
    "locations": [[6, 6]],
    "phases": [np.pi / 2],
    "relative_sf": False,
}
gabor_set = GaborSet(**gabor_params)
```

```python
canvas_size = [41, 41]
locations = [[20.0, 20.0]]  # fixed position
sizes_total = [10]  # variable size within range [5.0, 13.0]
sizes_center = [1.0]  # only center
sizes_surround = [1.1]  # doesn't matter, only center (has to be >1)
contrasts_center = [0.7]  # try 2 contrasts
contrasts_surround = [0.0]  # doesn't matter, only center
orientations_center = [np.pi/4, 3*np.pi / 4]  # 2 orientations
orientations_surround = [0.0]  # doesn't matter, center only
spatial_frequencies_center = [0.2]  # fixed spatial frequency
phases_center = [0.0]  # 2 center phases
grey_levels = [0.0]  # fixed grey level

# call center-surround
circular_grating = CenterSurround(
    canvas_size=canvas_size,
    locations=locations,
    sizes_total=sizes_total,
    sizes_center=sizes_center,
    sizes_surround=sizes_surround,
    contrasts_center=contrasts_center,
    contrasts_surround=contrasts_surround,
    orientations_center=orientations_center,
    orientations_surround=orientations_surround,
    spatial_frequencies_center=spatial_frequencies_center,
    phases_center=phases_center,
    grey_levels=grey_levels,
)

# plot the generated images
fig, axs = plt.subplots(1, 2, dpi=300)
for ax, img in zip(axs, circular_grating.images()):
    ax.imshow(img, cmap="gray", vmin=-1, vmax=1)
    ax.axis("off")

fig.savefig("choices.pdf", bbox_inches="tight", transparent=True)
```
