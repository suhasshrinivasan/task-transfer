```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# Seed for reproducibility
np.random.seed(42)

# Number of epochs
epochs = 100

# Generate a typical training loss that decreases over time
train_losses = np.linspace(1, 0.1, epochs) + np.random.normal(0, 0.05, epochs)

# Generate a typical validation loss that also decreases but might plateau or fluctuate
val_losses = np.linspace(1.2, 0.2, epochs) + np.random.normal(0, 0.1, epochs)

# Ensure losses don't go below zero
train_losses = np.maximum(train_losses, 0)
val_losses = np.maximum(val_losses, 0)

# Plot to visualize
fig, ax = plt.subplots(dpi=300)
ax.plot(train_losses, label="Training Loss", marker="o", linestyle="--", color="indianred")
# plot with discrete markers

ax.plot(val_losses, label="Validation Loss", marker="o", linestyle="--", color="teal")
ax.set_xlabel("Epochs")
ax.set_ylabel("Loss")
# ax.set_ylim(0, max(max(train_losses), max(val_losses)))
ax.set_yticks([])
ax.set_title("Simulated Training and Validation Losses")
ax.legend()
sns.despine(ax=ax, trim=True)
ax.spines["left"].set_visible(False)
print("Training Losses:\n", train_losses)
print("Validation Losses:\n", val_losses)
```
