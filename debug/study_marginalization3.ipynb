```python
data = pd.read_json("mc_error_comprehensive.json", orient="json")
# fig, ax = plt.subplots(dpi=300)

# Set the plot style
# sns.set
sns.set(style="whitegrid")

# Create a figure and axis
plt.figure(dpi=300)

# Swarm plot: No overlap of points
# sns.swarmplot(x="prior_dim", y="errors", data=data, dodge=True)

# Optional: Strip plot as an alternative (comment out if using swarmplot)
sns.stripplot(x="prior_dim", y="errors", data=data, jitter=False, dodge=True)

# Set labels and title
plt.xlabel("Prior Dimension")
plt.ylabel("Errors")
plt.title("Distribution of Errors for Each Prior Dimension")

# Display the plot
plt.show()
```
