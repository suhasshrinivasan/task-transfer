```python
import plotly.express as px
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_low_var = pd.read_json("check_bias_estimate_low_var.json", orient="json")
data_med_var = pd.read_json("check_bias_estimate_med_var.json", orient="json")
data = pd.read_json("check_bias_estimate.json", orient="json")
# fig, ax = plt.subplots(dpi=300)
# fig, ax = plt.subplots(dpi=300)

# Set the plot style
# sns.set
sns.set(style="whitegrid")

# Create a figure and axis
plt.figure(dpi=300)

# Swarm plot: No overlap of points
# sns.swarmplot(x="prior_dim", y="errors", data=data, dodge=True)

# Optional: Strip plot as an alternative (comment out if using swarmplot)
sns.stripplot(x="prior_dim", y="errors", data=data, jitter=False, dodge=True, color="blue")
sns.stripplot(x="prior_dim", y="errors", data=data_med_var, jitter=False, dodge=True, color="green")
sns.stripplot(x="prior_dim", y="errors", data=data_low_var, jitter=False, dodge=True, color="red")

# Set labels and title
plt.xlabel("Number of prior dimensions")
plt.ylabel("Errors")
plt.title("Errors for each prior dim size")
plt.xticks(np.arange(0, 46, 3))

# Display the plot
plt.show()
```
